## What is regression?

*Firstly*, I would tell you in Layman language: Regression is the task of predicting a value, like we do in any other supervised learning algorithm. <b>So, what is the difference between classification and regression?</b> In classification, we predict some discrete value, representative of some class label, but in regression we predict some continous value representative of some trend in the data.

If said in sophisticated way: In statistical modeling, regression analysis is a set of statistical processes for estimating the relationships between a dependent variable (often called the 'outcome variable') and one or more independent variables (often called 'predictors', 'covariates', or 'features').

### Let's understand that by example

#### generating artificial data

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_regression 
import random
```

```python
# NOISE

noise = []

for i in range(0, 50):
    x = round(random.uniform(0.0, 20.0), 2)
    noise.append(x)
    

# FEATURES    

X = np.random.randint(low=10, high=100, size=50)


# PREDICTED VALUES

m = 0.7
C = 0.3
Y = m*X + C

Y = Y.reshape(50,1)

li = []
for i in Y:
    for j in i:
        li.append(j)
        
Y = li

final_Y = []
for i in range(50):
    final_Y.append(Y[i] + noise[i])

Y = final_Y

plt.scatter(X, Y)
```

As it can be observed from above data that there is no ideal case scenario with data. Points are scattered but scattered in such a way that there is a correlation between X and Y. There is some noise in data, that data points are not ideally suiting y=mx+c.

### How to deal with such noise?

First of all consider each point on X as $x_1, x_2, x_3, \cdots , x_n$. For those points, there is corresponding $y_i$ and there is some noise $\epsilon $ for each point. How to deal with such noise? We consider that noise have gaussian distribution, with zero mean and $\sigma^2 $ variance. Let's visualize it:

```python
img = plt.imread('gaussian_noise.png')
plt.imshow(img)
```

$\epsilon $ will be between -1 to +1, because this region is highly probable region. What do I mean by that? Let's see:


```python
from scipy.stats import norm
```

```python
mu = 0
sigma = 4
x1 = 900
x2 = 1100
```

```python
z1 = ( x1 - mu ) / sigma
z2 = ( x2 - mu ) / sigma
```

```python
x = np.arange(z1, z2, 0.001) # range of x in spec
x_all = np.arange(-10, 10, 0.001) # entire range of x, both in and out of spec
# mean = 0, stddev = 1, since Z-transform was calculated
y = norm.pdf(x,0,1)
y2 = norm.pdf(x_all,0,1)
```

```python
fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('fivethirtyeight')
ax.plot(x_all,y2)

ax.fill_between(x,y,0, alpha=0.3, color='b')
ax.fill_between(x_all,y2,0, alpha=0.1)
ax.set_xlim([-4,4])
ax.set_xlabel('# of Standard Deviations Outside the Mean')
ax.set_yticklabels([])
ax.set_title('Normal Gaussian Curve')

plt.savefig('normal_curve.png', dpi=72, bbox_inches='tight')
plt.show()
```

Now that we assumed our noise gaussian, we are going to mathematically formulate our problem:

Using machine learning, we are trying to find the pattern or the relationship or the function governing that relationship between the response variable(dependent variable) and set of independent variables using a noisy data. 

$$\text{Our data is in the form:}\ \ \textbf{Y}^{}\ =\ f(x)\ +\ \epsilon $$
$$\ $$
$$\text{We must find}\ \hat{f}(x)\ \text{the best estimate of}\ f(x) \\ \text{so that we can predict values of Y because it is observable that}\ \epsilon \\ \text{is not a single value but one of the values from a gaussian distribution}$$
