```python
def train_cv_split(X,Y,cv_size):

    X_cv = X[int(cv_size*X.shape[0]):,:]

    Y_cv = Y[int(cv_size*Y.shape[0]):,:]

    X_train = X[0:int(cv_size*X.shape[0]),:]

    Y_train = Y[0:int(cv_size*Y.shape[0]),:]
    
    return X_train,X_cv,Y_train,Y_cv
```

```python
from sklearn.linear_model import Ridge
```

```python
from sklearn.metrics import mean_squared_error
```

```python
def estimate_mse_test(train_X,train_Y,cv_X,cv_Y):
    
    mse_test_estimates = []
    
    bootstrapped_samples = 100
    
    for _ in range(bootstrapped_samples):
        
        random_indices = np.random.choice(np.arange(0,train_X.shape[0]),size=train_X.shape[0])
        
        bootstrap_X_train = train_X[random_indices,:]
        
        bootstrap_Y_train = train_Y[random_indices,:]
        
        model_obj = Ridge(alpha=0.5)
        
        model_obj.fit(bootstrap_X_train,bootstrap_Y_train)
        
        Y_predict_cv = model_obj.predict(cv_X)
        
        mse_test_estimates.append(mean_squared_error(cv_Y,Y_predict_cv))
        
    mse_test_estimates = np.array(mse_test_estimates)
    
    return np.mean(mse_test_estimates),model_obj
```
