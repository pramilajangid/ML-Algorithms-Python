# LDA - as dimensionality reduction

## <font color=008080>Let's try to understand what is going on in our algorithm :</font>

## AIM

The aim of LDR is to provide us a vector, to project our data onto, such that BETWEEN GROUP variance should increase for each class and WITHIN GROUP variance should decrease.

#### Find such $\vec{w}$, such that,  when $\boldsymbol{X}$ is projected onto $\vec{w}$ which is a unit vector, within group variance of the data should be minimum and between group variance should be maximum, where $\boldsymbol{X}$ is our feature matrix.

For simplicity let's assume for now that we are classifying only for 2 classes.

Let's say for any $i^{th}$ example belonging to class '0' :

$$X^i\ =\ \begin{bmatrix}x^i_1\\x^i_2\\ \vdots \\ x^i_N\end{bmatrix}$$

$$\text{and,}\ \ \ \vec{w}\ =\ \begin{bmatrix}w_1\\w_2\\ \vdots \\w_N \end{bmatrix}$$

When we project ${X^{i}}$ onto $\vec{w}$, we get a new vector $\vec{p}$, such that:
$$\ $$
$$\vec{p} = \lambda_i \times \vec{w}$$

$$where, \ \lambda_i$$ is actually new data point after projection. And for all the data points after projection, means for respective class can be written as:

$$\mu^0_\lambda = \frac{\lambda^0_1 + \lambda^0_2 + \cdots + \lambda^0_{N_0}}{N_0}\ =\ \frac{\vec{w}^T.x^0_1 + \cdots +\vec{w}^T.x^0_{N_0}}{N_0}\ =\ \vec{w}^T\cdot \mu^0$$



$$\text{where,}\ \ \ \lambda_i = \vec{w}^T \cdot X_i\ ;\ \ \ \ \ \ \ \ \ \mu^0 = \begin{bmatrix} \mu^0_1 \\ \mu^0_2 \\ \vdots \\ \mu^0_N \end{bmatrix}\ \ \ \text{ is the mean vector of class '0' before projection}$$

$$\lambda^0_i = \vec{w}^T \cdot X^0_i\ \ \text{and}\ \ \lambda^1_i = \vec{w}^T \cdot X^1_i$$

Similarly variance after projection can be written as:

$$ var(\lambda^0)\ =\ \frac{\sum^{N_0}_{i=1}\left(\lambda^0_i - \mu^0_\lambda \right)^2}{N_0} = \frac{\sum^{N_0}_{i=1}\left(\vec{w}^T.X^0_i - \vec{w}^T.\mu^0 \right)^2 }{N_0}$$

$$ \Longrightarrow \frac{1}{N_0}\cdot \left[ \vec{w}^T.(X^0 - \mu^0) \right]\cdot \left[\vec{w}^T.(X^0 - \mu^0) \right]^T\ =\ \frac{1}{N_0}\cdot \left[\vec{w}^T.(X^0 - \mu^0).(X^0 - \mu^0)^T.\vec{w} \right] $$

### A. within group variance of the data

$$\boldsymbol{\sigma^2_{within}} = \frac{N_0 \cdot \sum^{N_0}_{i=1}\left(\lambda^0_i - \mu^0_\lambda \right)^2\ \boldsymbol{+}\ N_1 \cdot \sum^{N_1}_{i=1}\left(\lambda^1_i - \mu^1_\lambda \right)^2}{N_0\ +\ N_1}$$

where, 
- $N_0$ and $N_1$ are number of training example in '0' and '1' class respectively
- $\lambda^0$ and $\lambda^1$ are the values of $\lambda$ for each example in each class.
- $\mu^0_\lambda$ and $\mu^1_\lambda$ are the means of the projected data for class '0' and '1' respectively.

$$\boldsymbol{\sigma^2_{within}}\ =\ \vec{w}^T\cdot \boldsymbol{S_{within}}\cdot \vec{w}$$

where, $S_{within}$ is called within class scatter matrix.

$$\text{putting values for lambda as calculated above we get:}\\ \\ \boldsymbol{S_{within}} = \frac{N_0}{N_{train}}\cdot \left(X^0 - \mu^0 \right)\cdot \left(X^0 - \mu^0 \right)^T\ \ +\ \ \frac{N_1}{N_{train}}\cdot \left(X^1 - \mu^1 \right)\cdot \left(X^1 - \mu^1 \right)^T $$

### B. between group variance of the data

$$\boldsymbol{\sigma^2_{between}} = \frac{\sum^1_{k = 0} \left(\mu^i_\lambda - GM_\lambda \right)^2} {2}$$ 

where, 

- $GM_\lambda$ is the grand mean vector, after projecting data onto $\vec{w}$ :

$$GM_\lambda = \frac{\mu^0_\lambda \ +\ \mu^1_\lambda}{2} $$

$$ = \frac{\vec{w}^T\cdot \mu^0 + \vec{w}^T\cdot \mu^1 }{2}$$

$$\ $$
$$\text{Hence,}\ \ \ GM_\lambda = \vec{w}^T\cdot GM$$
$$\ $$

$$\text{where,}\ \ \ GM = \begin{bmatrix}GM_1 \\GM_2 \\ \vdots \\ GM_N \end{bmatrix} \begin{matrix} ..\text{grand mean for 1st feature vector} \\ ..\text{grand mean for 2nd feature vector}\\ \vdots \\ ..\text{grand mean for Nth feature vector} \end{matrix}$$

So, now $\sigma^2_{between}$ becomes:

$$\boldsymbol{\sigma^2_{between}} = \frac{\sum^1_{i=0} \left(\vec{w}^T\cdot \mu^i\ -\ \vec{w}^T\cdot GM\right)^2}{2}$$

Converting to vectorized form:

$$\mu \ =\ {\begin{bmatrix}\uparrow & \uparrow \\ \mu^0 & \mu^1 \\ \downarrow & \downarrow \end{bmatrix}}_{N\times 2}\ \ \ \ \ \ \ \ \ \ GM = {\begin{bmatrix}GM_1 \\ GM_2\\ \vdots \\ GM_N \end{bmatrix}}_{N\times 1}$$

$$\ $$

$$\mu - GM = {\begin{bmatrix}\uparrow & \uparrow \\ (\mu^0 - GM) & (\mu^1 - GM) \\ \downarrow & \downarrow \end{bmatrix}}_{N \times 2}$$

Hence $\sigma^2_{between}$ can be written as :

$$\boldsymbol{\sigma^2_{between}} = \vec{w}^T\cdot \boldsymbol{S_{between}}\cdot \vec{w}$$

$$\ $$

$$\text{where,}\ \ \ \boldsymbol{S_{between}} = \frac{(\mu - GM)\cdot (\mu - GM)^T}{2}$$

### Now our after projection within group variance should be minimum and between group variance should be maximum. So, let's define a cost function such that :

$$\textbf{Cost Function = }\ \ \ J(\vec{w})\ =\ \frac{\vec{w}^T\cdot S_{between}\cdot \vec{w}}{\vec{w}^T\cdot S_{within}\cdot \vec{w}}$$

$$\\ \text{should be maximum for given}\ \vec{w}$$

$$\text{That is:}\ \ \ \frac{\partial J(\vec{w})}{\partial w_1} = 0\ ;\ \ \ 
\frac{\partial J(\vec{w})}{\partial w_2} = 0\ ;\cdots\ ;\ \frac{\partial J(\vec{w})}{\partial w_N} = 0$$

For simplicity let's just consider:
$$ \vec{w} = \begin{bmatrix}w_1 \\ w_2 \end{bmatrix}\ \ \ \ \ \ \ \ \ \ S_{between} = \begin{bmatrix} \sigma^2_{{between}_1} & 0 \\ 0 & \sigma^2_{{between}_2} \end{bmatrix}\ \ \ \ \ \ \ \ \ \ S_{within} = \begin{bmatrix} \sigma^2_{{within}_1} & 0 \\ 0 & \sigma^2_{{within}_2} \end{bmatrix}$$

So the derivatives are: 

$$\frac{\partial J(\vec{w})}{\partial w_1}\ =\ 0\ \ \text{results in: }\ \ \ \frac{\sigma^2_{{between}_1}}{\sigma^2_{{within}_1}} = J(\vec{w})\cdot w_1$$
$$\ $$
$$\frac{\partial J(\vec{w})}{\partial w_2}\ =\ 0\ \ \text{results in: }\ \ \ \frac{\sigma^2_{{between}_2}}{\sigma^2_{{within}_2}} = J(\vec{w})\cdot w_2$$

Combining results:

$$\begin{bmatrix} \frac{\sigma^2_{{between}_1}}{\sigma^2_{{within}_1}} & 0 \\ 0 & \frac{\sigma^2_{{between}_2}}{\sigma^2_{{within}_2}} \end{bmatrix}\ \cdot \ \begin{bmatrix} w_1 \\ w_2 \end{bmatrix} \ =\ J(\vec{w})\cdot \begin{bmatrix}w_1 \\ w_2 \end{bmatrix}$$

$$ \boldsymbol{\left(S_{between}\cdot S^{\ \ -1}_{within} \right)\cdot \vec{w} = J\ (\vec{w})\cdot \vec{w} }$$

If observed, it is quiet clear that $J(\vec{w})$ will be one of eigen value of the matrix $\left(S_{between}\cdot S^{\ \ -1}_{within} \right)$. And since we have to maximize $J(\vec{w})$ we will take eigen vector with maximum eigen value.

### How to get that eigen vector? 

Singular Value Decomposition : np.linalg.svd($S_{between} \cdot S^{-1}_{within}$)
